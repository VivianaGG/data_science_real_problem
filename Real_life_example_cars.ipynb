#### Multicollinearity

```python
data_cleaned.columns.values
```

##### We expect a degree of multicollinearity but sklearn have not dedicated method to check this assumption, we must to use statsmodels through VIF (Variation inflation factor) -> produces an estimation of how much large the square root of the standard error of the estimate is compare with a situation when the variable is completelly uncorrelated with the other predictors.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year', 'EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

#### Year is very correlated with the others variables, so we drop it

```python
data_no_multicollinearity = data_cleaned.drop(['Year'],axis =1)
```

### Create the dummy variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first = True)
```

```python
data_with_dummies
```

### Rearrange a bit

```python
data_with_dummies.columns.values
```

```python
cols = ['Log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

### Linear regression model

#### Declare the input and the target

```python
targets = data_preprocessed['Log_price']
inputs = data_preprocessed.drop(['Log_price'], axis = 1)
```

#### Scale the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size = 0.2, random_state = 365)
```

train = to train the model
test = to test it

### Create the regression

```python
reg = LinearRegression()

reg.fit(x_train, y_train)
```

To obtain the predicted values, we use:

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size = 18)
plt.ylabel('predictions (y_hat)', size = 18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

Another popular model check is the residual = diff between the targets and predictions  

```python
sns.distplot(y_train - y_hat)
plt.title("Residual PDF", size = 18)
```

### R^2

```python
reg.score(x_train, y_train)
```

Our model explains 75 % of the variability of the data

#### Finding the weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns = ['Features'])
reg_summary['weights'] = reg.coef_
reg_summary
```

For a positive weight if the feature increases in value, so do the log_price and price
For a negative weight if the feature increases in value, the log_price and price decrease

```python
data_cleaned['Brand'].unique()
```

For the dummies variables as the dropped brand was Audi, it is the benchmark, so if the weight is positive shows that the brand checked is more expensive (BMW for instance) and the contrary if the weight is negative!...always compared with the dropped variable

   ### Testing

```python
plt.scatter(y_test, y_hat_test, alpha = 0.2)
plt.xlabel('Targets (y_test)', size = 18)
plt.ylabel('predictions (y_hat_test)', size = 18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```
